# **LIBRARIES**

```python
import pickle  # Mengimpor modul pickle untuk serialisasi dan deserialisasi objek Python
import pandas as pd  # Mengimpor pandas, pustaka untuk manipulasi data dan analisis
import numpy as np  # Mengimpor numpy, pustaka untuk operasi numerik
import ast  # Mengimpor ast, modul untuk memanipulasi dan mengurai ekspresi kode Python
```

# **MODEL INFERENCE**

## Model Loading

```python
with open('model_scaler.pkl', 'rb') as file_1:  # Membuka file 'model_scaler.pkl' dalam mode baca biner
    model_scaler = pickle.load(file_1)  # Memuat model_scaler dari file menggunakan pickle

with open('model_kmeans.pkl', 'rb') as file_2:  # Membuka file 'model_kmeans.pkl' dalam mode baca biner
    model_kmeans = pickle.load(file_2)  # Memuat model_kmeans dari file menggunakan pickle

with open('model_pca.pkl', 'rb') as file_3:  # Membuka file 'model_pca.pkl' dalam mode baca biner
    model_pca = pickle.load(file_3)  # Memuat model_pca dari file menggunakan pickle
```

## Creating Data Inference Dummy

```python
# Membuat 11 data dummy
np.random.seed(1)  # Menetapkan seed untuk generator angka acak agar hasilnya dapat direproduksi

# Menghasilkan 11 angka acak yang uniform untuk setiap variabel berikut:
BALANCE = np.random.uniform(0, 18495, 11)  # Saldo
BALANCE_FREQUENCY = np.random.uniform(0, 1, 11)  # Frekuensi saldo
PURCHASES = np.random.uniform(0, 41050, 11)  # Pembelian
ONEOFF_PURCHASES = np.random.uniform(0, 40624, 11)  # Pembelian sekali
INSTALLMENTS_PURCHASES = np.random.uniform(0, 14686, 11)  # Pembelian cicilan
CASH_ADVANCE = np.random.uniform(0, 47137, 11)  # Uang muka tunai
PURCHASES_FREQUENCY = np.random.uniform(0, 1, 11)  # Frekuensi pembelian
ONEOFF_PURCHASES_FREQUENCY = np.random.uniform(0, 1, 11)  # Frekuensi pembelian sekali
PURCHASES_INSTALLMENTS_FREQUENCY = np.random.uniform(0, 1, 11)  # Frekuensi pembelian cicilan
CASH_ADVANCE_FREQUENCY = np.random.uniform(0, 1.5, 11)  # Frekuensi uang muka tunai
CASH_ADVANCE_TRX = np.random.uniform(0, 123, 11)  # Transaksi uang muka tunai
PURCHASES_TRX = np.random.uniform(0, 347, 11)  # Transaksi pembelian
CREDIT_LIMIT = np.random.uniform(50, 30000, 11)  # Batas kredit
PAYMENTS = np.random.uniform(0.049513, 50721, 11)  # Pembayaran
MINIMUM_PAYMENTS = np.random.uniform(0.019163, 76406, 11)  # Pembayaran minimum
PRC_FULL_PAYMENT = np.random.uniform(0, 1, 11)  # Persentase pembayaran penuh
TENURE = np.random.choice(['6', '7', '8', '9', '10', '11', '12'], 11)  # Masa jabatan

# Membuat DataFrame Inference
data_inf = pd.DataFrame({  # Membuat DataFrame menggunakan pandas
    'NAME' : 'Dendi',
    'BALANCE': BALANCE,  # Menetapkan kolom 'BALANCE' dengan data BALANCE
    'BALANCE_FREQUENCY': BALANCE_FREQUENCY,  # Menetapkan kolom 'BALANCE_FREQUENCY' dengan data BALANCE_FREQUENCY
    'PURCHASES': PURCHASES,  # Menetapkan kolom 'PURCHASES' dengan data PURCHASES
    'ONEOFF_PURCHASES': ONEOFF_PURCHASES,  # Menetapkan kolom 'ONEOFF_PURCHASES' dengan data ONEOFF_PURCHASES
    'INSTALLMENTS_PURCHASES': INSTALLMENTS_PURCHASES,  # Menetapkan kolom 'INSTALLMENTS_PURCHASES' dengan data INSTALLMENTS_PURCHASES
    'CASH_ADVANCE': CASH_ADVANCE,  # Menetapkan kolom 'CASH_ADVANCE' dengan data CASH_ADVANCE
    'PURCHASES_FREQUENCY': PURCHASES_FREQUENCY,  # Menetapkan kolom 'PURCHASES_FREQUENCY' dengan data PURCHASES_FREQUENCY
    'ONEOFF_PURCHASES_FREQUENCY': ONEOFF_PURCHASES_FREQUENCY,  # Menetapkan kolom 'ONEOFF_PURCHASES_FREQUENCY' dengan data ONEOFF_PURCHASES_FREQUENCY
    'PURCHASES_INSTALLMENTS_FREQUENCY': PURCHASES_INSTALLMENTS_FREQUENCY,  # Menetapkan kolom 'PURCHASES_INSTALLMENTS_FREQUENCY' dengan data PURCHASES_INSTALLMENTS_FREQUENCY
    'CASH_ADVANCE_FREQUENCY': CASH_ADVANCE_FREQUENCY,  # Menetapkan kolom 'CASH_ADVANCE_FREQUENCY' dengan data CASH_ADVANCE_FREQUENCY
    'CASH_ADVANCE_TRX': CASH_ADVANCE_TRX,  # Menetapkan kolom 'CASH_ADVANCE_TRX' dengan data CASH_ADVANCE_TRX
    'PURCHASES_TRX': PURCHASES_TRX,  # Menetapkan kolom 'PURCHASES_TRX' dengan data PURCHASES_TRX
    'CREDIT_LIMIT': CREDIT_LIMIT,  # Menetapkan kolom 'CREDIT_LIMIT' dengan data CREDIT_LIMIT
    'PAYMENTS': PAYMENTS,  # Menetapkan kolom 'PAYMENTS' dengan data PAYMENTS
    'MINIMUM_PAYMENTS': MINIMUM_PAYMENTS,  # Menetapkan kolom 'MINIMUM_PAYMENTS' dengan data MINIMUM_PAYMENTS
    'PRC_FULL_PAYMENT': PRC_FULL_PAYMENT,  # Menetapkan kolom 'PRC_FULL_PAYMENT' dengan data PRC_FULL_PAYMENT
    'TENURE': TENURE  # Menetapkan kolom 'TENURE' dengan data TENURE
})

data_inf  # Menampilkan DataFrame
```

## SCALING

```python
# Menghapus kolom 'TENURE' dari DataFrame dan menyimpannya sebagai data_inf_num
data_inf_num = data_inf.drop(columns=['NAME'])
```

```python
# Menerapkan skala pada data_inf_num menggunakan model_scaler dan menyimpannya sebagai scaled_inf
scaled_inf = model_scaler.transform(data_inf_num)
```

## PCA

```python
pca_inf = model_pca.transform(scaled_inf)  # Menerapkan PCA (Principal Component Analysis) pada scaled_inf menggunakan model_pca dan menyimpannya sebagai pca_inf
pca_inf  # Menampilkan hasil transformasi PCA
```

## PREDICTING

```python
# Prediksi pada Inference-Set

cluster_inf = model_kmeans.predict(pca_inf)  # Memprediksi cluster untuk data pca_inf menggunakan model_kmeans dan menyimpannya sebagai cluster_inf
cluster_inf  # Menampilkan hasil prediksi cluster
```

```python
# Konversi ke dalam DataFrame

cluster_inf_df = pd.DataFrame(cluster_inf, columns=['Cluster - Prediction'])  # Membuat DataFrame dari cluster_inf dengan kolom 'Cluster - Prediction'
cluster_inf_df  # Menampilkan DataFrame hasil prediksi cluster
```

```python
# Menggabungkan Inference-Set dan Prediksi

tabel_inference = pd.concat([data_inf, cluster_inf_df], axis=1)  # Menggabungkan data_inf dan cluster_inf_df secara horizontal (axis=1) dan menyimpannya sebagai tabel_inference
tabel_inference  # Menampilkan tabel_inference hasil penggabungan
```
